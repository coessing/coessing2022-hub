Ū, V̄, u and v are vectors that can be plotted and inspected. Let's look at them in detail! (probably just map the speed instead of the individual velocity components)

Let's create a "continent mask" to overlay over our velocity field. In this case we want to have a mask that has all the elements for which `bathymetry > 0` (over sea level) equal to `NaN`s (these will be our continents) and all elements for which `bathymetry < 0` (the ocean) equal to `0`. This is easy to do since, in julia, we can index an array with a conditional.

```julia
mask = zeros(Nx, Ny)
mask[ bathymetry .> 0 ] .= NaN
mask[ bathymetry .< 0 ] .= 0.0
```

Our masked velocity field will just be `Um + mask` (remember that `Um` is a 3D array)

```julia
# Mean zonal velocity
U = Um[:, :, 1] .+ mask

fig = Figure(resolution = (1000, 500))
ax  = Axis(fig[1, 1])
heatmap!(ax, U, colorrange = (-0.2, 0.2))
current_figure()
```

The same thing can be done for the meridional velocity `V`. Note that `U` and `V` do not have the same dimensions! 

```julia
# Mean meridional velocity
V = Vm[:, 1:end-1, 1] .+ mask

fig = Figure(resolution = (1000, 500))
ax  = Axis(fig[1, 1])
heatmap!(ax, V, colorrange = (-0.1, 0.1))
current_figure()
```
